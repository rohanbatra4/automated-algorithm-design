```python
file_train_data = "../input/fashionmnist/fashion-mnist_train.csv"
file_test_data = "../input/fashionmnist/fashion-mnist_test.csv"
df_train = pd.read_csv(file_train_data)
df_test = pd.read_csv(file_test_data)
df_train.head()
```

```python
def get_features_labels(df):
    features = df.values[:, 1:]/255
    labels = df['label'].values
    return features, labels
```

```python
train_features, train_labels = get_features_labels(df_train)
test_features, test_labels = get_features_labels(df_test)
```

```python
print(train_features.shape)
print(test_features.shape)
plt.figure()
_ = plt.imshow(np.reshape(train_features[221, :], (28,28)), 'gray')
```

```python
#one hotting
train_labels = tf.keras.utils.to_categorical(train_labels)
test_labels = tf.keras.utils.to_categorical(test_labels)
```

```python
train_labels.shape
```

```python
train_labels[255]
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(30, activation=tf.nn.relu, input_shape=(784,)))
model.add(tf.keras.layers.Dense(20, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(10, activation=tf.nn.softmax))

model.compile(loss='categorical_crossentropy', optimizer='rmsprop',metrics=['accuracy'])
```

```python
model.summary()
```

```python
EPOCHS = 3 #Gradient descent is iderative process! Epoch = 2 feeds data through twice.
# Too many epochs = overfitting, too few = underfitting
BATCH_SIZE = 128 #how much to put into training at once
model.fit(train_features, train_labels, epochs=EPOCHS, batch_size=BATCH_SIZE)
```

```python
test_loss, test_acc = model.evaluate(test_features, test_labels)
```
